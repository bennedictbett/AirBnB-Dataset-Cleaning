```python
import kagglehub
 
path = kagglehub.dataset_download("arianazmoudeh/airbnbopendata")
print("Original dataset location:", path)
```

```python
import shutil, os

project_folder = os.getcwd()
new_path = os.path.join(project_folder, "airbnb_dataset")

if not os.path.exists(new_path):
    shutil.copytree(path, new_path)
    print("Dataset copied to:", new_path)
else:
    print("Dataset already exists in project folder.")
```

```python
import pandas as pd     
```

```python
df = pd.read_csv(new_path + "/Airbnb_Open_Data.csv")
```

```python
# Define columns to keep and drop
columns_to_keep = ['NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
columns_to_drop = ['reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license', 'id']
```

```python
len(columns_to_keep)                                
```

```python
len(columns_to_drop)
```

```python
df.columns
```

```python
# Drop unnecessary columns
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.shape
```

```python
df
```

```python
# Clean column names
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df
```

```python
df.duplicated().sum()
```

```python
# remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df.drop(columns=['last_review'], inplace=True)
```

```python
# drop nan values
df.dropna(inplace=True)
```

```python
df['host_identity_verified'] = df['host_identity_verified'].str.upper()
```

```python
df['instant_bookable'].apply(lambda x: 1 if x== True else 0)


```

```python
df['instant_bookable'] = df['instant_bookable'].apply(lambda x: 1 if x== True else 0)
```

```python
df.reset_index(inplace=True)
```

```python
type(df['instant_bookable'][15])
```

```python
type(df['price'][12])
```

```python
df['price'].head(3)
```

```python
# Replace dollar sign with empty string
df['price'] = df['price'].str.replace('$','')
```

```python
# Replace comma with empty string
df['price'] = df['price'].str.replace(',','')
```

```python
# Remove spaces from the price column
df['price'] = df['price'].str.replace(' ','')
```

```python
# Convert price column to integer type
df['price'] = df['price'].astype(int)
```

```python
df.drop(columns=['index'], inplace=True)
```

```python
df
```

```python
# Save the cleaned DataFrame to a new CSV file without the index
df.to_csv(new_path + "/cleaned_airbnb_data.csv", index=False)
```
